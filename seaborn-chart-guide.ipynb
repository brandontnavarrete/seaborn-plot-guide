# Basic Plot Types:

## Line Plot (sns.lineplot())

A line plot in Seaborn is a type of plot used to display the relationship between a continuous dependent variable and one or more continuous independent variables. It connects the data points in a line, making it easy to visualize trends and patterns in the data.

In Seaborn, line plots can be created using the lineplot function. 
Here's an example of how to create a basic line plot in Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
sns.lineplot(x=x, y=y)

# Show the plot
plt.show()
```

<div class="alert alert-block alert-success">
<b>lineplot:</b> This code will generate a line plot that shows 
the relationship between x and y. 
You can customize the appearance of the plot using various arguments of the lineplot function, 
such as changing the color and style of the line.
</div>

# Customizing Plots:

## Changing color, size, and shape of markers

* Customizing the color, size, and shape of markers in Seaborn plots can be done using various arguments in the plotting function. Here's an overview of how to do it:

`Color` : The color of markers can be specified using the color argument in the plotting function. You can specify a single color, a list of colors, or a color palette from Seaborn. For example, to set the color of markers to blue, you can use the following code:

```python
sns.scatterplot(x=x, y=y, color="blue")
```

`Size`: The size of markers can be specified using the size argument in the plotting function. You can specify a single size or a list of sizes that corresponds to the data points. For example, to set the size of markers to 10, you can use the following code:


```python
sns.scatterplot(x=x, y=y, size=10)
```

`Shape`: The shape of markers can be specified using the marker argument in the plotting function. You can specify a single shape or a list of shapes that corresponds to the data points. Some common shapes are "o" for circles, "s" for squares, and "+" for plus signs. For example, to set the shape of markers to squares, you can use the following code:


```python
sns.scatterplot(x=x, y=y, marker="s")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Plot the data with square markers
sns.scatterplot(x=x, y=y, marker="s")

# Show the plot
plt.show()
```

## Adding titles, x- and y-labels, and legends

<div class="alert alert-block alert-success">
<b>titles, labels, legends:</b> Setting titles, x and y labels, and providing a legend will help guide your audience to your conclusion rather than they make their own.

</div>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x1 = np.random.normal(0, 1, 100)
y1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(1, 1, 100)
y2 = np.random.normal(1, 1, 100)

# Plot the data
# Note the TWO plots
sns.scatterplot(x=x1, y=y1, label="Data 1")
sns.scatterplot(x=x2, y=y2, label="Data 2")

# Add plot title
plt.title("Scatter Plot of X and Y")

# Add x-axis label
plt.xlabel("X")

# Add y-axis label
plt.ylabel("Y")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

<div class="alert alert-block alert-warning">
<b> Note :</b> In a Seaborn plot, the legend may not be necessary if you only have one plot, but if you have multiple plots or lines on the same plot, you may want to add a legend to differentiate between them. 
</div>

## Setting axis limits

<div class="alert alert-block alert-success">
<b>axis limits:</b> Setting axis limits in Seaborn (and Matplotlib) can be useful to control the range of values that are displayed on the x-axis and y-axis. This can be useful in cases where you want to focus on a specific region of the data or where you want to standardize the axis limits across multiple plots.
</div>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Plot the data
sns.scatterplot(x=x, y=y)

# Set the x-axis and y-axis limits
plt.xlim(-3, 3)
plt.ylim(-3, 3)

# Show the plot
plt.show()
```

```python
* In this example, notice the range of the x axis and the y axis. They match the values we have predefined.
```

## Using different color palettes

<div class="alert alert-block alert-success">
<b>color palettes:</b> Seaborn provides a variety of color palettes that can be used in your plots. To use a different color palette, you can pass a string representing the palette name to the palette argument when plotting. Here's an example of how to use different color palettes in a Seam
    </div>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([1, 2, 3, 4, 5])

# Plot the data using a default palette
sns.barplot(x=x, y=y)

# Add plot title
plt.title("Bar Plot using Default Palette")

# Show the plot
# Note plt.show() creates a plot point and I repeated below the instance.
plt.show()

# Plot the data using a different palette
sns.barplot(x=x, y=y, palette="husl")

# Add plot title
plt.title("Bar Plot using HUSL Palette")

# Show the plot
plt.show()
```


<div class="alert alert-block alert-success">
<b>color palettes:</b> You can also use a custom color palette by passing a list of colors to the palette argument. The list should contain as many colors as there are unique categories in your plot. Here's an example:
</div>


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([1, 2, 3, 4, 5])

# Define a custom color palette
palette = ["teal", "skyblue", "silver", "gray", "black"]

# Plot the data using the custom palette
sns.barplot(x=x, y=y, palette=palette)

# Add plot title
plt.title("Bar Plot using Custom Palette")

# Show the plot
plt.show()
```

# Plotting Data Types:

Plots can be categorized into two main types based on the type of data they represent:   
continuous data and categorical data. Plotting with continuous data involves mapping continuous variables to the x and y axes and visualizing the relationship between them.  

On the other hand, plotting with categorical data involves mapping categorical variables to the x and y axes and visualizing the distribution of values within each category.

## Categorical

When plotting with categorical data, it's important to understand the underlying distribution of values within each category and how these values are related to one another. This can help you identify patterns, relationships, and trends in your data 

## Continous

In contrast, when plotting with continuous data, it's important to understand the relationship between the variables and how they change over time. This can help you identify patterns and trends in your data, such as linear or non-linear relationships

## Summary

In summary, both plotting with categorical and continuous data have their own importance and can provide valuable insights into your data. The choice of which type of plot to use depends on the nature of your data and the questions you want to answer.

# Plotting with Categorical / Discrete/ Data:

## Count Plot (sns.countplot())

<div class="alert alert-block alert-success">
<b>countplot:</b> A count plot is a type of categorical plot used in Seaborn to visualize the distribution of the number of occurrences of each category in a categorical variable. It is essentially a bar plot where the height of each bar represents the count of the number of occurrences of a specific category.

</div>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
fruit_stand = ["apple", "banana", "strawberry", "apple", "banana", "banana", "apple", "cherry", "cherry", "apple"]

# Create the count plot
sns.countplot(x=fruit_stand)

# Add a title to the plot
plt.title("Fruit Stand Sales")

# Show the plot
plt.show()
```


<div class="alert alert-block alert-success">
<b>countplot:</b> This will create a count plot of the distribution of fruit sales at the fruit stand, with each bar representing the count of one type of fruit. The x-axis shows the different types of fruit, and the y-axis shows the count of each type. 
</div>


## Box Plot (sns.boxplot())

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a sample dataset
np.random.seed(0)
data = {"Category": ["A", "B", "C", "A", "B", "C"],
        "Value": np.random.normal(100, 20, 6)}
df = pd.DataFrame(data)

# Create the box plot
sns.boxplot(x="Category", y="Value", data=df)

# Add a title to the plot
plt.title("Box Plot of Categorical Data")

# Show the plot
plt.show()
```

## Violin Plot (sns.violinplot())

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset('tips')

# Create a violin plot of total bill amounts by day of the week
sns.violinplot(x='day', y='total_bill', data=tips)

# Add a title to the plot
plt.title('Distribution of Total Bill Amounts by Day of the Week')

# Add x and y labels
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')

# Display the plot
plt.show()
```


<div class="alert alert-block alert-success">
<b>violintplot:</b> In this example, we first load the tips dataset from the Seaborn library. We then create a violin plot of the total_bill column, grouped by the day column. The plot displays the distribution of total bill amounts for each day of the week. Finally, we add a title, x and y labels, and display the plot.</div>


## Swarm Plot (sns.swarmplot())

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset('tips')

# Create a swarm plot of total bill amounts by day of the week
sns.swarmplot(x='day', y='total_bill', data=tips)

# Add a title to the plot
plt.title('Distribution of Total Bill Amounts by Day of the Week')

# Add x and y labels
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')

# Display the plot
plt.show()
```

# Plotting with Multi-Variate Data:

## Pair Plot (sns.pairplot())

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
iris = sns.load_dataset('iris')

# Create a pairplot of the iris dataset
sns.pairplot(iris, hue='species', palette='husl')

# Add a title to the plot
plt.suptitle('Pairplot of the Iris Dataset', fontsize=16, y=1.05)

# Display the plot
plt.show()
```

## Joint Plot (sns.jointplot())

```python
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset('tips')

# Create a joint plot of total bill and tip amount
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')

# Add a title to the plot
plt.suptitle('Joint Plot of Total Bill and Tip Amount', fontsize=16, y=1.05)

# Display the plot
plt.show()
```

## Heatmap (sns.heatmap())

<div class="alert alert-block alert-warning">
<b>Note:</b>  A heatmap will pair up with a correlation matrix
</div>



```python
import seaborn as sns
import pandas as pd
import numpy as np

# generate a sample correlation matrix
data = np.random.randn(100, 5)
corr = np.corrcoef(data.T)

# plot only the lower triangle of the matrix
              # color           # value print # format    # masking the top 
sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".2f", mask=np.triu(np.ones_like(corr, dtype=bool)))
```

* the closer to 1 values indicates a stronger postitive correlation and the correlation and the closer to 0 values indicate a negative correlation

## Faceting multiple plots with FacetGrid


<div class="alert alert-block alert-success">
<b>facetgrid:</b> Faceting is a way to display multiple plots in a single figure, where each plot represents a subset of the data. In Seaborn, the FacetGrid is used to implement faceting.

For example, you can create a faceted plot for different categories of a categorical variable or for different ranges of a continuous variable.

```python
import seaborn as sns
import pandas as pd
import numpy as np

# generate a sample data set
data = sns.load_dataset("titanic")

# create a faceted scatter plot
g = sns.FacetGrid(data, col="class")
g.map(sns.scatterplot, "age", "fare")
```
